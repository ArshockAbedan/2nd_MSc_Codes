We are creating masked images for using in EdgeConnect

```python
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
mask_path = "/home/aabedjoo/Aref_project/night2day_missing/qd_imd/test/"
Dataset_root = "/home/aabedjoo/Aref_project/Main_Dataset/night/test/"
edgeconnect_root = "/home/aabedjoo/Aref_project/edge-connect/examples/night2day/M1-A1/"
dsize = (256, 256)

onlyfiles_night = [f for f in listdir(Dataset_root) if isfile(join(Dataset_root, f))]
i = 1
for file_name_night in onlyfiles_night:
    # Load the original image
    file_night = Dataset_root + file_name_night 
    image_night = cv2.imread(file_night)

    file_name_arr = file_name_night.split('.')
    destination_file_name = file_name_arr[0]
    ############################
    # a mask
    a_number = np.random.randint(0,9999)
    number_str = str(a_number)
    zero_filled_number = number_str.zfill(5)
    mask_img_file_name = mask_path + zero_filled_number + '_test.png'

    # Using 0 to read image in grayscale mode
    mask_img = cv2.imread(mask_img_file_name,0)
    resized_mask_img = cv2.resize(mask_img, dsize, interpolation = cv2.INTER_AREA)
    
    ####################################
    rows,cols = resized_mask_img.shape
    for row in range(rows):
        for col in range(cols):
            if resized_mask_img[row,col] == 0 :
                image_night[row,col] = 255
                resized_mask_img[row,col] = 255
            elif resized_mask_img[row,col] == 255:
                resized_mask_img[row,col] = 0
    
    destination_file_path = edgeconnect_root + "images/" + destination_file_name + ".png"
    destination_mask_path = edgeconnect_root + "masks/" + destination_file_name + ".png"

    # Saving the image
    cv2.imwrite(destination_file_path, image_night)
    cv2.imwrite(destination_mask_path, resized_mask_img)
    i =  i + 1
    print(i)
```
